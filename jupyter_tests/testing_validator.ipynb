```python
import pandas as pd
```

```python
mydata = pd.DataFrame(
    {'year': range(1900, 2017)}
)
china = mydata.copy()
usa = mydata.copy()
uk = mydata.copy()
canada = mydata.copy()
china['country'] = 'China'
china['gdp'] = mydata['year'] * 2 + 1000
usa['country'] = 'USA'
usa['gdp'] = mydata['year'] * 3 + 1000
uk['country'] = 'UK'
uk['gdp'] = mydata['year'] + 100
canada['country'] = 'Canada'
canada['gdp'] = mydata['year'] * 1.5 + 200
alldata = pd.concat([china, usa])
alldata.reset_index(drop=True, inplace=True)
```

```python
import numpy as np
```

```python
data = pd.get_dummies(alldata)
data['country'] = alldata['country']
```

```python
from sklearn.model_selection import GroupKFold, KFold
```

```python
from ravenclaw.regression import Validator, Regressor
```

```python
data.tail()
```

```python
from ravenclaw.preprocessing import PolynomialTransformer
```

```python
validator = Validator(folds=KFold(n_splits=10), regressor=Regressor(name='simple', polynomial_model=PolynomialTransformer(degree=2)))
```

```python
X_training = data.drop(axis=1, labels=['gdp', 'country'])
y_training = data['gdp']
```

```python
validator.validate(X=X_training, y=y_training, groups=data['country'])
```

```python
reg = validator._trained_regressors[0]
```

```python
reg.coefficients
```
