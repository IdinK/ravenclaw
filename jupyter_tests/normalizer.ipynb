```python
from polarbear.preprocessing import Normalizer
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10, 3))
df[0] = df[0]*100 + 10
df[1] = df[1]-100
display(df)
```

```python
normalizer = Normalizer()
normalizer.normalize(df, inplace=True)
display(df)
```

```python
df2 = pd.DataFrame(np.random.randn(10, 3))
df2[0] = df2[0]*90 + 10
df2[1] = df2[1]-50
display(df2)

display(normalizer.normalize(df2))
```
